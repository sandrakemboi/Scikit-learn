```python
import sklearn
```

```python
from sklearn.datasets import load_iris
```

```python
load_iris()
```

```python
load_iris(return_X_y =True) 
```

```python
inputval,outputval = load_iris(return_X_y = True)
print(outputval)
```

```python
from sklearn.linear_model import LinearRegression#Here we are importing the model we want to use;models are just python objects that can be used to create machine learning models.
```

```python
#Assign model as variable
model = LinearRegression() #Here have  created a model.
```

```python
model.fit(inputval,outputval)
```

```python
model.predict(inputval)
```

```python
#KNN modeµ
from sklearn.neighbors import KNeighborsRegressor
```

```python
model2 = KNeighborsRegressor()
```

```python
model2.fit(inputval,outputval)
```

```python
predicted = model2.predict(inputval)
print(predicted)
```

```python
import matplotlib.pyplot as plt
plt.scatter(predicted,outputval)
```
